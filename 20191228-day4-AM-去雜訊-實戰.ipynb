```python
%matplotlib inline
import matplotlib.pyplot as plt
cherry = plt.imread('cherry.jpg')
print(cherry.shape)
plt.imshow(cherry)
plt.show()
```

```python
#  
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
cherry_bilateral = denoise_bilateral(image= cherry, multichannel= True)
plt.imshow(cherry_bilateral)
plt.imsave("cherry_bilateral.jpg", cherry_bilateral)
plt.show()
```

```python
#  
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
cherry_tv = denoise_tv_chambolle(image= cherry, weight= 0.05, multichannel= True)
plt.imshow(cherry_tv)
plt.imsave("cherry_tv.jpg", cherry_tv)
plt.show()
```
