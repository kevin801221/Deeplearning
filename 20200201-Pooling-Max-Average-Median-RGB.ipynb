```python
import numpy as np
pool = (4, 3)
def avg_pooling(img, pool):        
    max_height= int(img.shape[0]//pool[0])
    max_width = int(img.shape[1]//pool[1])
    pool_img = np.zeros((max_height, max_width))
    for ii in range(max_height):
        for jj in range(max_width):
                    pool_img[ii, jj] = np.mean(img[ii*pool[0]:ii*pool[0]+pool[0], jj*pool[1]:jj*pool[1]+pool[0]])
    return(pool_img)

def max_pooling(img, pool):        
    max_height= int(img.shape[0]//pool[0])
    max_width = int(img.shape[1]//pool[1])
    pool_img = np.zeros((max_height, max_width))
    for ii in range(max_height):
        for jj in range(max_width):
                    pool_img[ii, jj] = np.max(img[ii*pool[0]:ii*pool[0]+pool[0], jj*pool[1]:jj*pool[1]+pool[0]])
    return(pool_img)

def med_pooling(img, pool):        
    max_height= int(img.shape[0]//pool[0])
    max_width = int(img.shape[1]//pool[1])
    pool_img = np.zeros((max_height, max_width))
    for ii in range(max_height):
        for jj in range(max_width):
                    pool_img[ii, jj] = np.median(img[ii*pool[0]:ii*pool[0]+pool[0], jj*pool[1]:jj*pool[1]+pool[0]])
    return(pool_img)
```

```python
import matplotlib.pyplot as plt
andy = plt.imread('andy3.jpg')
andy_result = np.zeros((int(andy.shape[0]//pool[0]), int(andy.shape[1]//pool[1]), 3)).astype('uint8')
andy_result[:, :, 0] = avg_pooling(andy[:, :, 0], pool)
andy_result[:, :, 1] = avg_pooling(andy[:, :, 1], pool)
andy_result[:, :, 2] = avg_pooling(andy[:, :, 2], pool)
andy.shape, andy_result.shape
```
