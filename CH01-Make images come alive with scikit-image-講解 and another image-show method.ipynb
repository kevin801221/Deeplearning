## skimage庫的全名為scikit-image Scikit， 是對scipy.ndimage進行了擴展，提供了更多的圖片處理理功能。
## skimage包含很多的子模型，各個子模型具有不同的功能，如下

#### io:讀取,保存和顯示圖片和視訊
#### color:顏色空間變換
#### data:提供一些测試圖片和樣本數據
#### filters:圖像增强，邊緣檢測，排序濾波器，自動擴值
#### draw:操作於numpy數組上的基本圖形繪製，包括線條，矩陣
#### transform:幾何變換和其他變換，如旋轉，拉伸和變換等
#### exposure:图像强度调整，例如，直方图均衡化等
#### feature:特征检测和提取， 例如，纹理分析等
#### graph:图论操作，例如，最短路径
#### measure图像属性测量，例如，相似度和轮廓
#### morphology形态学操作，如开闭运算，骨架提取等
#### novice简化的用于教学目的的接口
#### restoration:修复算法，例如去卷积算法，去噪等
#### segmentation图像分割为多个区域

```python
%matplotlib inline

from skimage import io
import matplotlib.pyplot as plt
Gray_Sample = io.imread('./train/1000_c.png', as_gray=True)
print(Gray_Sample.shape)  #(128, 176)

plt.figure(figsize=(3,3))
plt.imshow(Gray_Sample, cmap='gray')
plt.title('Sample_c')
#plt.axis('on')
plt.axis('off')
plt.show()
```

```python
# 讀入劉德華 相片
andy = plt.imread('andy.jpg')
plt.imshow(andy)
plt.show()
```

```python
# 使用color.rgb2gray轉換成gray
from skimage import color
plt.imshow(color.rgb2gray(andy), cmap='gray')
plt.show()
```

```python
# 驗證轉換成為灰階之後 再轉為彩色
from skimage import color
plt.imshow( color.gray2rgb( color.rgb2gray(andy) )  )
print(color.gray2rgb( color.rgb2gray(andy) ).shape)
plt.show()
```
