```python
# morphology
# 讓圖變胖 變瘦 

# 建議轉為grayscale
from skimage import io  #也可以用matplotlib把圖輸出來
andy = io.imread('andy.jpg', as_gray = True) # as_gray = True 會讓數值變小
andy
#andy[:3]
```

```python
import matplotlib.pyplot as plt
plt.imread('andy.jpg')[:3]
```

```python
plt.imshow(andy, cmap='gray')
plt.show()
```

```python
# 以下是使用 threshold_otsu 搭配 erosion , dilation
andy_gray = andy * 255.0
from skimage.filters import threshold_otsu
thresh = threshold_otsu(andy_gray)
print(thresh)
andy_bin_otsu = andy_gray > thresh
plt.imshow(andy_bin_otsu, cmap='gray')
plt.show()
```

```python
# 讓圖形  erosion 變細 變瘦
from skimage import morphology 
selem = morphology.rectangle(4, 4)
eroded_andy = morphology.binary_erosion(andy_bin_otsu, selem = selem)
plt.imshow(eroded_andy, cmap='gray')
plt.show()
```

```python
# 讓圖形  dilation + otsu 變細 變瘦
from skimage import morphology 
selem = morphology.rectangle(4, 4)
eroded_andy = morphology.binary_dilation(andy_bin_otsu)
plt.imshow(eroded_andy, cmap='gray')
plt.show()
```

```python
# 以下是使用 threshold_local 搭配 erosion , dilation
```

```python
andy_gray = andy * 255.0
from skimage.filters import threshold_local
thresh = threshold_local(andy_gray, 35, offset=5)
#print(thresh)
andy_bin_local = andy_gray > thresh
plt.imshow(andy_bin_local, cmap='gray')
plt.show()
```

```python
#erosion
from skimage import morphology
selem = morphology.rectangle(4, 4)
eroded_andy = morphology.binary_erosion(andy_bin_local, selem = selem)
plt.imshow(eroded_andy, cmap='gray')
plt.show()
```

```python
#dilation
from skimage import morphology
selem = morphology.rectangle(4, 4)
eroded_andy = morphology.binary_dilation(andy_bin_local)
plt.imshow(eroded_andy, cmap='gray')
plt.show()
```
