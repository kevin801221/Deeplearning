```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
andy = plt.imread('andy.jpg')

plt.imshow(andy)
plt.show()
```

```python
from skimage.segmentation import slic
from skimage.color import label2rgb   #功能：转换标记矩阵到RGB图像


segments = slic(andy, n_segments=8000, sigma=3)

#andy_segmented = label2rgb(segments, andy, kind='overlay')
andy_segmented = label2rgb(segments, andy, kind='avg')
plt.imshow(andy_segmented)
plt.show()
```
