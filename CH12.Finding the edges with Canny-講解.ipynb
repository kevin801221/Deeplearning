```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
andy = plt.imread('andy.jpg')

plt.imshow(andy)
plt.show()
```

```python
# 找邊界 filters.sobel
from skimage import color
from skimage.filters import sobel
andy_gray = color.rgb2gray(andy)
andy_gray_sobel = sobel(andy_gray)
plt.imshow(andy_gray_sobel, cmap='gray')
plt.show()
```

```python
# 找邊界 feature.canny
from skimage import color
from skimage.feature import canny
andy_gray = color.rgb2gray(andy)
andy_gray_canny = canny(andy_gray, sigma=2)
plt.imshow(andy_gray_canny, cmap='gray')
plt.show()
```
